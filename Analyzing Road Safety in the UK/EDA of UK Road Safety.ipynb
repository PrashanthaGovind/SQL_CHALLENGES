**Meaningful Insights**

- The number of casualties and accidents severity is high when the vehicle speed is more than 60.
- More number of accidents recorded in Urban area.
- Maximum number of accidents caused by the "car" and least accidents caused by the "Train".
- MOre number of accidents caused by the "Right Hand Drive" vehicle.
